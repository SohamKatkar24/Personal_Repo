```python
from sklearn import datasets
import numpy as np

iris = datasets.load_iris()
X = iris.data[:, [2, 3]]
y = iris.target

print('Class labels:', np.unique(y))

```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.get_cmap('viridis',3))
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Iris Data: Petal Width vs. Petal Length')
plt.colorbar(label='Species')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

#Split the data as 70% training and 30% as test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print("Training set (X, y):", X_train.shape, y_train.shape)
print("Testing set (X, y):", X_test.shape, y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

#using K-nearest neighbour classifier
knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(X_train, y_train)

# Predictions on the testing data
y_pred = knn.predict(X_test)
print(y_pred)

# Plotting testing data with predictions
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred, cmap=plt.cm.get_cmap('viridis', 3))
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Testing Data with Predicted Labels')
plt.colorbar(label='Predicted Species')
plt.show()
```
